```python
def preprocess(images, labels, is_mnist=True):

    print("\nBefore Processing:")
    print("Shape:", images.shape)
    print("Range:", images.min(), images.max())

    # Normalize
    images = images.astype("float32") / 255.0

    # Reshape MNIST
    if is_mnist:
        images = np.expand_dims(images, axis=-1)

    # One-hot encode
    labels = tf.keras.utils.to_categorical(labels, 10)

    print("\nAfter Processing:")
    print("Shape:", images.shape)
    print("Range:", images.min(), images.max())

    return images, labels

# Apply
x_train_mnist_p, y_train_mnist_p = preprocess(x_train_mnist, y_train_mnist, True)
x_train_cifar_p, y_train_cifar_p = preprocess(x_train_cifar, y_train_cifar, False)
```

#Problem 4 — Code (Data Augmentation)

```python
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.1),
    tf.keras.layers.RandomZoom(0.1)
])

indices = np.random.choice(len(x_train_cifar), 5)

fig, axes = plt.subplots(5, 4, figsize=(10,12))

for i, idx in enumerate(indices):
    original = x_train_cifar[idx]

    # ORIGINAL IMAGE
    axes[i,0].imshow(original.astype("uint8"))
    axes[i,0].set_title("Original")
    axes[i,0].axis('off')

    for j in range(1,4):
        # Normalize only for augmentation
        img = original.astype("float32")

        aug = data_augmentation(tf.expand_dims(img, 0), training=True)
        aug = aug.numpy()[0]

        # 🔥 FINAL FIX: convert back to uint8
        aug = np.clip(aug * 255, 0, 255).astype("uint8")

        axes[i,j].imshow(aug)
        axes[i,j].set_title("Augmented")
        axes[i,j].axis('off')

plt.tight_layout()
plt.savefig("augmentation_demo.png", dpi=300)
plt.show()
```

#Problem 1 — Manual Conv2D (NumPy only)


```python
import numpy as np

def conv2d(image, kernel, stride=1, padding=0):
    # Add padding
    if padding > 0:
        image = np.pad(image, ((padding, padding), (padding, padding)), mode='constant')

    H, W = image.shape
    kH, kW = kernel.shape

    # Output size
    out_H = (H - kH) // stride + 1
    out_W = (W - kW) // stride + 1

    output = np.zeros((out_H, out_W))

    # Convolution
    for i in range(out_H):
        for j in range(out_W):
            patch = image[i*stride:i*stride+kH, j*stride:j*stride+kW]
            output[i, j] = np.sum(patch * kernel)

    return output


# Test Image
image = np.array([
    [3,1,0,2,4],
    [1,5,3,2,1],
    [0,2,6,4,3],
    [2,3,1,5,2],
    [1,0,2,3,4]
])

# Sobel-X kernel
kernel = np.array([
    [-1,0,1],
    [-2,0,2],
    [-1,0,1]
])

output = conv2d(image, kernel, stride=1, padding=0)

print("Output Feature Map:\n", output)
print("Output Shape:", output.shape)
```

The conv2d function manually implements convolution using NumPy.

Steps:
1. Zero padding is applied if required
2. Kernel is slid across the image using nested loops
3. Element-wise multiplication is performed
4. Results are summed to produce output feature map

This replicates how CNN convolution layers work internally.

#Problem 3 — LeNet-5 Implementation

```python
import tensorflow as tf
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Conv2D(6, (5,5), activation='tanh', input_shape=(28,28,1)),
    layers.AveragePooling2D(pool_size=(2,2), strides=2),

    layers.Conv2D(16, (5,5), activation='tanh'),
    layers.AveragePooling2D(pool_size=(2,2), strides=2),

    layers.Flatten(),
    layers.Dense(120, activation='tanh'),
    layers.Dense(84, activation='tanh'),
    layers.Dense(10, activation='softmax')
])

model.summary()
```

Parameter Calculation (First Conv Layer)
Formula:
(K × K × C_in + 1) × C_out

= (5 × 5 × 1 + 1) × 6
= (25 + 1) × 6
= 156 parameters


LeNet-5 uses Average Pooling instead of Max Pooling because:

- It was designed in early CNN research where smoothing was preferred
- Average pooling preserves background information

Modern CNNs prefer MaxPooling because:
- It captures strongest features
- Helps in better feature extraction

#Problem 4 — Custom CNN (CIFAR-10)


Architecture (ASCII)
Input (32×32×3)
↓
Conv → BN → ReLU → MaxPool
↓
Conv → BN → ReLU → MaxPool
↓
Conv → BN → ReLU → MaxPool
↓
GlobalAveragePooling
↓
Dense → Dropout → Dense(10)

```python
model = models.Sequential([

    # Block 1
    layers.Conv2D(32, (3,3), padding='same', input_shape=(32,32,3)),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D(),

    # Block 2
    layers.Conv2D(64, (3,3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D(),

    # Block 3
    layers.Conv2D(128, (3,3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D(),

    # Head
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])

model.summary()
```

#Design Rationale
The architecture uses three convolutional blocks to progressively extract features.

Batch Normalisation stabilises training and speeds convergence.
ReLU introduces non-linearity.

GlobalAveragePooling reduces parameters and prevents overfitting.

Dropout is added to reduce overfitting in the classification head.

The model maintains a balance between complexity and generalisation.

#Analysis Questions
#Q1
Two 3×3 layers:
Params = 2 × (3×3×C×F) = 18CF

One 5×5 layer:
Params = 25CF

Thus, two 3×3 layers use fewer parameters.

Advantages:
- More non-linearity
- Better feature extraction

#Q2
Batch Normalisation normalises inputs to each layer.

It is applied before activation.

Benefits:
1. Faster convergence
2. Reduces internal covariate shift
3. Allows higher learning rates

#Q3
GlobalAveragePooling reduces each feature map to a single value.

This reduces parameters and overfitting.

Replacing with Flatten:
- Increases parameters significantly
- May cause overfitting
- Retains spatial information but less efficient

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import random

# Seed
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)

# Load MNIST
from tensorflow.keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Preprocess
x_train = x_train.astype("float32")
x_test = x_test.astype("float32")

x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)

y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(6, (5,5), activation='tanh', input_shape=(28,28,1)),
    tf.keras.layers.AveragePooling2D(pool_size=(2,2), strides=2),

    tf.keras.layers.Conv2D(16, (5,5), activation='tanh'),
    tf.keras.layers.AveragePooling2D(pool_size=(2,2), strides=2),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(120, activation='tanh'),
    tf.keras.layers.Dense(84, activation='tanh'),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(x_train, y_train,
                    epochs=15,
                    batch_size=64,
                    validation_split=0.1)

# Test accuracy
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test Accuracy:", test_acc)
```

```python
# Loss plot
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.savefig("lenet_sgd_curves.png")
plt.show()

# Accuracy plot
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training vs Validation Accuracy")
plt.legend()
plt.savefig("lenet_sgd_curves.png")
plt.show()
```

#Problem 4 — Regularisation Experiment

```python
def build_model(use_dropout=False, use_bn=False):
    model = tf.keras.Sequential()

    for filters in [32,64]:
        model.add(tf.keras.layers.Conv2D(filters,(3,3),padding='same',input_shape=(32,32,3)))
        if use_bn:
            model.add(tf.keras.layers.BatchNormalization())
        model.add(tf.keras.layers.Activation('relu'))
        model.add(tf.keras.layers.MaxPooling2D())
        if use_dropout:
            model.add(tf.keras.layers.Dropout(0.3))

    model.add(tf.keras.layers.Flatten())
    if use_dropout:
        model.add(tf.keras.layers.Dropout(0.5))

    model.add(tf.keras.layers.Dense(10,activation='softmax'))

    return model
```

Dropout reduces overfitting by randomly disabling neurons.
Batch Normalisation stabilises training.

Best performance is achieved when both are used together.

```python
import tensorflow as tf
import numpy as np

def grad_cam(model, image, class_index, last_conv_layer_name):


    grad_model = tf.keras.models.Model(
        inputs=model.input,
        outputs=[
            model.get_layer(last_conv_layer_name).output,
            model.output
        ]
    )

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(image)
        loss = predictions[:, class_index]

    grads = tape.gradient(loss, conv_outputs)

    pooled_grads = tf.reduce_mean(grads, axis=(0,1,2))

    conv_outputs = conv_outputs[0]

    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)

    heatmap = tf.maximum(heatmap, 0)
    heatmap /= tf.reduce_max(heatmap) + 1e-8

    return heatmap.numpy()
```

```python
from tensorflow.keras import layers, Model, Input

inputs = Input(shape=(28,28,1))

x = layers.Conv2D(6,(5,5),activation='tanh')(inputs)
x = layers.AveragePooling2D(pool_size=(2,2), strides=2)(x)

x = layers.Conv2D(16,(5,5),activation='tanh')(x)
x = layers.AveragePooling2D(pool_size=(2,2), strides=2)(x)

x = layers.Flatten()(x)
x = layers.Dense(120,activation='tanh')(x)
x = layers.Dense(84,activation='tanh')(x)

outputs = layers.Dense(10,activation='softmax')(x)

model = Model(inputs, outputs)
```

```python
# Sample
img = x_test[0]
img_tensor = np.expand_dims(img, axis=0)

# Prediction
pred = model.predict(img_tensor)
class_idx = np.argmax(pred)


last_conv_layer_name = None
for layer in reversed(model.layers):
    if 'conv' in layer.name:
        last_conv_layer_name = layer.name
        break

# Grad-CAM
heatmap = grad_cam(model, img_tensor, class_idx, last_conv_layer_name)

# Resize
import cv2
heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))

# Plot
plt.imshow(img.squeeze(), cmap='gray')
plt.imshow(heatmap, cmap='jet', alpha=0.5)
plt.axis('off')
plt.savefig("gradcam_results.png")
plt.show()
```

#Problem 1 — Feature Extraction (Frozen Base)

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Seed
tf.random.set_seed(42)
np.random.seed(42)

# Load CIFAR-10
from tensorflow.keras.datasets import cifar10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Reduce data (avoid crash)
x_train = x_train[:10000]
y_train = y_train[:10000]

x_test = x_test[:2000]
y_test = y_test[:2000]

# One-hot encoding
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

# Data pipeline
def preprocess(image, label):
    image = tf.image.resize(image, (96,96))
    image = image / 255.0
    return image, label

train_ds = tf.data.Dataset.from_tensor_slices((x_train, y_train)).map(preprocess).batch(32)
val_ds = tf.data.Dataset.from_tensor_slices((x_test, y_test)).map(preprocess).batch(32)

# Load VGG16
base_model = tf.keras.applications.VGG16(
    weights='imagenet',
    include_top=False,
    input_shape=(96,96,3)
)

# Freeze base
for layer in base_model.layers:
    layer.trainable = False

# Custom head
x = base_model.output
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dense(256, activation='relu')(x)
x = tf.keras.layers.Dropout(0.5)(x)
outputs = tf.keras.layers.Dense(10, activation='softmax')(x)

model = tf.keras.Model(inputs=base_model.input, outputs=outputs)

# Compile
model.compile(
    optimizer=tf.keras.optimizers.Adam(0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train
history = model.fit(train_ds, epochs=10, validation_data=val_ds)

# Plot
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Frozen Model")
plt.legend()
plt.savefig("tl_frozen.png")
plt.show()
```

The pre-trained VGG16 model is used as a fixed feature extractor.

All convolutional layers are frozen, and only the custom classification head is trained.

This allows faster training and prevents overfitting when data is limited.

#Problem 2 — Fine-Tuning

```python

for layer in base_model.layers[-4:]:
    layer.trainable = True

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

early_stop = tf.keras.callbacks.EarlyStopping(
    patience=5, restore_best_weights=True
)


history_ft = model.fit(
    train_ds,
    epochs=10,
    validation_data=val_ds,
    callbacks=[early_stop]
)

# Plot combined
plt.plot(history.history['val_accuracy'], label='Frozen')
plt.plot(history_ft.history['val_accuracy'], label='Fine-tuned')
plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy")
plt.title("Fine-Tuning")
plt.legend()
plt.savefig("tl_finetuned.png")
plt.show()
```

A smaller learning rate is used during fine-tuning to avoid destroying pre-trained weights.

Only higher layers are unfrozen since they capture task-specific features.

#Problem 3 — Ablation Study

```python
def run_experiment(unfreeze_layers):


    for layer in base_model.layers:
        layer.trainable = False


    for layer in base_model.layers[-unfreeze_layers:]:
        layer.trainable = True

    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-5),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )


    small_train = train_ds.take(100)
    small_val = val_ds.take(20)

    hist = model.fit(
        small_train,
        epochs=3,
        validation_data=small_val,
        verbose=0
    )

    return hist.history['val_accuracy'][-1]


results = {
    "Top 2 layers": run_experiment(2),
    "Top 8 layers": run_experiment(8),
    "All layers": run_experiment(len(base_model.layers))
}

print("Ablation Results:", results)
```

Unfreezing more layers increases model flexibility but also increases risk of overfitting.

Lower layers contain general features, while higher layers are task-specific.
